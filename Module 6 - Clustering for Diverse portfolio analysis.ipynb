```python
table = table.pct_change()*100
table.dropna(axis=0, inplace=True)
table.head()
```

## Query 6.2

Calculate average annual percentage return and volatility of all 30 stocks over a theoretical one year period

```python
annual_return = table.mean()*252
annual_return
```

```python
import math
volatility = table.std()*math.sqrt(252)
volatility
```

```python
# annual_return = annual_return.to_numpy()
# volatility = volatility.as_matrix()
X = np.vstack((annual_return, volatility)).T
X
```

## Query 6.3
Cluster the 30 stocks according to their mean annual Volatilities and Returns using K-means clustering. Identify the optimum number of clusters using the Elbow curve method

```python
from sklearn.cluster import KMeans
% matplotlib inline
import matplotlib.pyplot as plt

score=[]
for k in range(2,15):
    kmeans = KMeans(n_clusters=k)
    score.append(kmeans.fit(X).score(X))
    
plt.plot(range(2,15), score)
plt.grid(True)
plt.show()
    
```

According to the elbow curve, we deduce that the number of optimum clusters can be 5

```python
kmeans = KMeans(n_clusters=5).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)

plt.figure(figsize=(10,10))
plt.scatter(X[:,0], X[:,1], c= kmeans.labels_.astype(float), cmap='RdYlBu')
# plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
for i, txt in enumerate(table.columns):
    plt.annotate(txt, (X[i,0], X[i,1]))
plt.show()
```

## Query 6.4
Prepare a separate Data frame to show which stocks belong to the same cluster

```python
group = pd.DataFrame(table.columns, columns=['Stock'])
group['Cluster'] = kmeans.labels_
group
```
