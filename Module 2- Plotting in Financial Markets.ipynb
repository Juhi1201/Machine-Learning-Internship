## Query 2.6

```python
adani = pd.read_csv('ADANIPOWER.csv')
castrol = pd.read_csv('CASTROLIND.csv')
gmr = pd.read_csv('GMRINFRA.csv')
godrej = pd.read_csv('GODREJIND.csv')
mindtree = pd.read_csv('MINDTREE.csv')

adani = adani[adani['Series']=='EQ']
gmr = gmr[gmr['Series']=='EQ']
godrej = godrej[godrej['Series']=='EQ']
castrol = castrol[castrol['Series']=='EQ']
mindtree = mindtree[mindtree['Series']=='EQ']
```

```python
adani = adani['Close Price']
castrol = castrol['Close Price']
gmr = gmr['Close Price']
godrej = godrej['Close Price']
mindtree = mindtree['Close Price']
jublfood = train2['Close Price']
```

```python
newdf = pd.concat([adani.rename('adani'),castrol.rename('castrol'),
                   gmr.rename('gmr'),godrej.rename('godrej'),mindtree.rename('mindtree'),jublfood.rename('jublfood')],axis=1)
newdf.head()
```

```python
newdf_pct = newdf.pct_change()*100
newdf_pct.drop(newdf_pct.index[0],inplace=True)
newdf_pct.head()
```

```python
sns.pairplot(newdf_pct)
```

## Query 2.7

```python
newdf_pct['adani_ra'] = newdf_pct['adani'].rolling(window=7,center=False).std()
newdf_pct['jublfood_ra'] = newdf_pct['jublfood'].rolling(window=7,center=False).std()
newdf_pct.head(20)
```

## Query 2.8

```python
nifty = pd.read_csv('Nifty50.csv')
nifty.head()
```

```python
nifty_cp = nifty['Close']
nifty_cp = nifty_cp.pct_change()*100
nifty_cp.drop(nifty_cp.index[0],inplace=True)
nifty_cp.columns=['Nifty']
nifty_cp.head()
```

```python
nifty_pct = nifty_cp.rolling(window=7,center=False).std()
train2['Beta'] = nifty_pct
train2.head()
```

```python
plt.figure(figsize=(20,10))
nifty_pct.plot(kind='line',color='r')
newdf_pct['adani_ra'].plot(kind='line',color='b')
newdf_pct['jublfood_ra'].plot(kind='line',color='g')
plt.show()
```

## Query 2.9

```python
buysell = train2['Average Price']
train2['Sma'] = buysell.rolling(window=21,center=False).mean()
train2['Lma'] = buysell.rolling(window=34,center=False).mean()
```

```python
plt.figure(figsize=(20,10))
buysell.plot(kind='line',color='r')
train2['Sma'].plot(kind='line',color='b')
train2['Lma'].plot(kind='line',color='g')
plt.show()
```

## Query 2.10

```python
train2['bollinger'] = buysell.rolling(window=14,center=False).mean()
std = buysell.rolling(window=14,center=False).std()
train2['upper'] = train2['bollinger'] + (std*2)
train2['lower'] = train2['bollinger'] - (std*2)

plt.figure(figsize=(20,10))
buysell.plot(kind='line',color='black')
train2['upper'].plot(kind='line',color='g')
train2['lower'].plot(kind='line',color='r')
train2['bollinger'].plot(kind='line',color='b')
plt.show()
```
